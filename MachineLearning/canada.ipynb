```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 
```

```python
df=pd.read_csv('canada.csv')
df.head(3)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Income Per Capita in USSD')
plt.scatter(df.year,df.percapitaincome, marker = '+', color='purple')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.percapitaincome)
```

```python
reg.predict([[2020]])
```
